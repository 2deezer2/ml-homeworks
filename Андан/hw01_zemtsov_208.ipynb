### Задача 1 (Парадокс инспектора)

Автобусы отходят от автостанции с 8:00 до 20:00. Первый автобус отходит ровно в 8:00. Затем интервалы между автобусами случайны, независимы и равновероятно равны либо 5-и, либо 10-и минутам. Будем считать, что за минуту на автостанцию приходит ровно один пассажир, и все пассажиры едут ближайшим автобусом.

Проведите $10^4$ экспериментов и с их помощью:

__а) [5 баллов]__ Постройте гистограмму количества автобусов, отошедших от автостанции за сутки. Похоже ли визуально распределение на нормальное?

__б) [10 баллов]__ Инспектор Тимон выбирает равновероятно один из всех автобусов отошедших от автостанции за сутки. Постройте гистограмму числа пассажиров на этом автобусе. Оцените математическое ожидание и дисперсию этого числа.

__в) [10 баллов]__ Инспектор Пумба приходит на автостанцию в случайный момент времени, равномерный от 8:00 до 20:00 и садится в первый пришедший автобус. Постройте гистограмму числа пассажиров на этом автобусе. Оцените математическое ожидание и дисперсию этого числа.

__г) [10 баллов]__ Как изменятся ответы на эти вопросы, если время между автобусами будет экспоненциально со средним в 10 минут?

__Пояснения:__ первый автобус можно не включать в выборку, Пумбу за пассажира можно не считать. 

```python
# your code here ( ⚆ ω ⚆)
# Этот вариант с тупым циклом, но он быстрее.
import numpy as np
import random
from matplotlib import pyplot as pyp
random.seed('kek')
# Вычитал, что зерно может быть практически любым типом данных
n = 10**4
# Циклом для глупых
array = np.empty((n,1))
for i in range(n):
  bus = 0
  time = np.array([8, 0])
  while time[0] < 20:
    time[1] += random.choice([5,10])
    if time[1]>=60:
      time[1]=time[1]%60
      time[0]+=1
      if time[0]==20:
        break
    bus+=1
  array[i] = bus
pyp.hist(array, bins = 30)
pyp.grid(visible = True)
pyp.title("Количество автобусов по дням")
pyp.xlabel("Количество автобусов за день")
pyp.ylabel("Количество дней")
pyp.show()

# Можем заметить, что распределение лишь отдалённо напоминает нормальное, возможно, это потому, что гистограмма объединяет промежутки значений аргумента.
```

```python
# your code here ( ⚆ ω ⚆)
# Этот вариант с умным циклом, но он блин 9 сек выполняется, это наверное из-за конкатенате, но к сожалению, я не смог вычитать ничего быстрее, что выполняет такую задачу
import numpy as np
import random
from matplotlib import pyplot as pyp
random.seed('kiss')
# Вычитал, что зерно может быть практически любым типом данных
n = 10**4
kekis = np.empty((n,1))
avtobusik = np.empty((n,1))
for i in range(n):
  #Создаю массив минимальной длины, который может быть из промежутков времени => длина массива - количество автобусов за день
  array = np.random.choice([5,10], size = 12*6)
  while True:
    plus = np.random.choice([5,10], size = 1)
    if np.sum(array)+plus<=720 :
      array = np.concatenate((array, plus))
    else:
      break
  avtobusik[i] = random.choice(array)
  kekis[i] = len(array)
pyp.hist(kekis, bins = 24)
pyp.grid(visible = True)
pyp.title("Количество автобусов по дням")
pyp.xlabel("Количество автобусов за день")
pyp.ylabel("Количество дней")
pyp.show()
```

### Задача 2 (Парадокс Хуана Мануэля Родригеса Паррондо)

У Атоса, Портоса и Арамиса по 1000 франков. Атос постоянно ходит в казино А, где каждый раз выигрывает один франк с вероятностью $0.49$ и проигрывает один франк с вероятностью $0.51$. Портос ходит в казино Б, где ситуация интереснее :) 

- Если богатство посетителя кратно трём, то посетитель выигрывает франк с вероятностью $0.09$ и проигрывает один франк с вероятностью $0.91$
- Если богатство посетителя не кратно трём, то посетитель выигрывает франк с вероятностью $0.74$ и проигрывает один франк с вероятностью $0.26$

Арамис каждый раз выбирает казино А или казино Б равновероятно. Проведите $10^4$ симуляций эволюции благосостояния Атоса, Портоса и Арамиса.Каждая симуляция предусматривает $1000$ посещений казино. Все три игрока все деньги носят с собой и ни на что не тратят :)

__а) [10 баллов]__ Постройте на графике $10$ случайных траекторий изменений богатства Атоса, по горизонтали — номер посещени казино, по вертикали — богатство. На том же графике постройте усреднённую по всем экспериментам тракторию изменения богатства.

__б) [10 баллов]__ Постройте на графике $10$ случайных траекторий изменений богатства Портоса, по горизонтали —  номер посещени казино, по вертикали — богатство. На том же графике постройте усреднённую по всем экспериментам тракторию изменения богатства.

__в) [10 баллов]__ Постройте на графике $10$ случайных траекторий изменений богатства Арамиса, по горизонтали — номер посещени казино, по вертикали — богатство. На том же графике постройте усреднённую по всем экспериментам тракторию изменения богатства.
    

```python
# your code here  =^･ω･^=
import random
import numpy as np
random.seed("kekis")
n = 10**4
#Казино А легко симулируется такой штукой:(-1)**(np.random.binomial(n = 1, p=0.49)+1)
#Создаём казино B
def Casino_B(money):
    if money%3==0:
        return((-1)**(np.random.binomial(n = 1, p=0.09)+1))
    else:
        return((-1)**(np.random.binomial(n = 1, p=0.74)+1))
#Начинаем симулировать
Atos_simulations = np.empty((n, 1000))
Partos_simulations = np.empty((n, 1000))
Aramis_simulations = np.empty((n, 1000))
for i in range(n):
    #Симулируем Атоса
    Atos_money = 1000
    for el in range(1000):
        Atos_money += (-1)**(np.random.binomial(n = 1, p=0.49)+1)
        Atos_simulations[i][el] = Atos_money
    #Симулируем Партоса
    Partos_money = 1000
    for el in range(1000):
        Partos_money += Casino_B(Partos_money)
        Partos_simulations[i][el] = Partos_money
    #Симулируем Арамиса
    Aramis_money = 1000
    for el in range(1000):
        Aramis_money += random.choice(((-1)**(np.random.binomial(n = 1, p=0.49)+1), Casino_B(Aramis_money)))
        Aramis_simulations[i][el] = Aramis_money
# Выбираем Рандомные варианты богатства Атоса
kek = random.sample(range(10000), k = 10)
for el in sorted(kek):
    pyp.plot(Atos_simulations[el], label = f"simulation №{el}")
pyp.plot(np.average(Atos_simulations, axis = 0), label = f"Среднее богатство")
pyp.grid(visible = True)
pyp.legend()
pyp.title("Богатство Атоса")
pyp.xlabel("Номер посещения казино")
pyp.ylabel("Количество денег")
pyp.show()
```

```python
# То же самое с Партосом
kek = random.sample(range(10000), k = 10)
for el in sorted(kek):
    pyp.plot(Partos_simulations[el], label = f"simulation №{el}")
pyp.plot(np.average(Partos_simulations, axis = 0), label = f"Среднее богатство")
pyp.grid(visible = True)
pyp.legend()
pyp.title("Богатство Партоса")
pyp.xlabel("Номер посещения казино")
pyp.ylabel("Количество денег")
pyp.show()
```

```python
# То же самое с Арамисом
kek = random.sample(range(10000), k = 10)
for el in sorted(kek):
    pyp.plot(Aramis_simulations[el], label = f"simulation №{el}")
pyp.plot(np.average(Atos_simulations, axis = 0), label = f"Среднее богатство")
pyp.grid(visible = True)
pyp.legend()
pyp.title("Богатство Арамиса")
pyp.xlabel("Номер посещения казино")
pyp.ylabel("Количество денег")
pyp.show()
```

### Задача 3 (Парадокс Берксона)

Предположим, что результаты ЕГЭ школьников по русскому и математике независимы и хорошо аппроксимируются 
нормальным распределением с ожиданием 60 и стандартным отклонением 10. 

УШЭ (Урюпинская Школа Экономики) ныне престижна и забирает себе всех школьников набравших более $n$ баллов в сумме по русскому и математике. 

Для каждого $n$ от $80$ до $160$ с шагом $5$ случайно создайте $10^4$ школьников и поделите их на прошедших и не прошедших в УШЭ.


__а) [5 баллов]__ Постройте график числа прошедших в УШЭ в зависимости от $n$. 

__б) [5 баллов]__ Постройте график выборочной корреляции между результатами по русскому и математике в зависимости от $n$ среди прошедших в УШЭ.

__в) [5 баллов]__ Постройте график выборочной корреляции между результатами по русскому и математике в зависимости от $n$ среди не прошедших в УШЭ.

```python
# your code here  (ﾉ´ з `)ノ
import numpy
import random
import matplotlib.pyplot as plt
random.seed("kekis")
shkol = 10**4
balls = np.empty((2, shkol))
# создаю школьников
balls[0] = np.random.normal(60.0, 10.0, size = shkol) # Предмет 1
balls[1] = np.random.normal(60.0, 10.0, size = shkol) # Предмет 2
n = np.arange(80, 160, 5) # ряд баллов n.
Summa = np.sum(balls, axis=0)
itog = np.empty(len(n)) #Итоговый массив для пункта
for el in n:
    itog[el//5-16] = len(Summa[Summa >= el])
itog = np.array(list(map(int, itog)))
plt.plot(n, itog)
pyp.grid(visible = True)
pyp.title("Количество прошедших в УШЭ")
pyp.xlabel("Проходной балл")
pyp.ylabel("Количество прошедших студентов")
pyp.show()
# Вид графика - отрицателньный корень нечётной степени больше 1, он имеет такой вид, так как сначала каждое
# увеличение проходного отсеивает мало людей, после количество отсеянных растёт, позже снова падает.
```

```python
Dataset = np.empty(len(n))
for el in n:
    kek = np.where(Summa>=el)
    Dataset[el//5-16] = np.corrcoef(balls[0, kek], balls[1, kek])[1][0]
pyp.plot(n, Dataset)
pyp.grid(visible = True)
pyp.title("Выборочная корреляция по русскому и математике среди прошедших")
pyp.xlabel("Проходной балл")
pyp.ylabel("Выборочная корреляция")
pyp.show()
# Тут мы видим как раз таки парадокс, когда мы ставим некоторое условие на 2 штуки.
```
